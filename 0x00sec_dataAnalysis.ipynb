```python
data = pd.read_csv("0x00secUserData.csv")
data["Number of Posts in Thread"] = 0
new_df = data.groupby(['Thread ID'])['Number of Posts in Thread'].agg('count').reset_index()
average_numeber_posts_per_thread = new_df['Number of Posts in Thread'].mean()
new_df = new_df.sort_values(by='Number of Posts in Thread', ascending=False)
display(average_numeber_posts_per_thread)
top_twentyFive = new_df.head(50)
display(top_twentyFive)
x_axis = new_df['Number of Posts in Thread']
  
# Calculating mean and standard deviation
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)
  
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```
